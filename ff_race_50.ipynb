```python
import pandas as pd
```

```python
from geopy.geocoders import Nominatim
```

```python
from geopy.exc import GeocoderTimedOut

```

```python
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv('/Users/soe.sone/Desktop/Running_data_project/ff_race_50.csv')
```

```python
df.head()
```

```python
df2 = df.dropna(axis=1)
```

```python
df2.head()
```

```python
df2.loc[:, 'Time'] = pd.to_timedelta(df2['Time'])
```

```python
df2['Total_minutes'] = df2['Time'].dt.total_seconds() / 60
```

```python
df2['Total_minutes'] = df2['Total_minutes'].round().astype(int)
```

```python
df2.head()
```

```python
df2.rename(columns={'Division' : 'Gender'}, inplace=True)
```

```python
def get_lat_long(city, state): 
    address = f"{city}, {state}"
    try: 
        geolocator = Nominatim(user_agent = 'running', timeout=10)
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None 
    except GeocoderTimedOut:
        return None, None
```

```python
df2['latitue'],df2['longitude'] = zip(*df2.apply(lambda x: get_lat_long(x['City'], x['State']), axis = 1))
```

```python
df2.head()
```

```python
df2['latlong'] = df2['latitue'].astype(str) + ', ' + df2['longitude'].astype(str)
```

```python
df2.head()
```

```python
df2.rename(columns={'latitue' : 'latitude'}, inplace=True)
```

```python
df2.to_csv('ultracleanedupdate_output.csv', index = False)
```
